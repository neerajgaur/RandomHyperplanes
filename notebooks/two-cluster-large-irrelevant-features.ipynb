```python
import numpy as np
import matplotlib.pyplot as plt
from random_hyperplanes import simulations
```

```python
N_ESTIMATORS = 100
SCORE_AT = 97.5

num_points = 1000
num_features = 2
contamination = 0.05

X, X_anon = simulations.two_cluster_simulation_data(
    contamination=contamination,
    num_points=num_points, 
    num_features=num_features,
    variance=0.01,
    mean=5.0,
    scale='large'
)

simulations.X_LIM = [-6.0, 6.0]
simulations.Y_LIM = [np.min(X[:, 1]) - 1.0, np.max(X[:, 1]) + 1.0]
```

```python
plt.scatter(X[:, 0], X[:, 1], c='b', marker='o', alpha=0.5)
plt.scatter(X_anon[:, 0], X_anon[:, 1], c='r', marker='x', alpha=0.5)
plt.xlim(simulations.X_LIM)
plt.ylim(simulations.Y_LIM)
plt.show()
```

```python
simulations.run_comparison_simul(X=X, X_anon=X_anon, n_estimators=N_ESTIMATORS, score_at=SCORE_AT)
```
